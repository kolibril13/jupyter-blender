# Move the default cube

A minimal demo that the in-process kernel can see `bpy`. Make sure the default cube is present in the scene, then run the cells below.


```python
import bpy
import ipywidgets as widgets
from IPython.display import display
```

```python
cube = bpy.data.objects.get("Cube")
cube
```

```python
x = widgets.FloatSlider(min=-3.0, max=3.0, step=0.1, value=0.0, description="x")
y = widgets.FloatSlider(min=-3.0, max=3.0, step=0.1, value=0.0, description="y")
z = widgets.FloatSlider(min=-3.0, max=3.0, step=0.1, value=0.0, description="z")

def on_change(_change):
    if cube is not None:
        cube.location = (x.value, y.value, z.value)

for w in (x, y, z):
    w.observe(on_change, names="value")

display(widgets.VBox([x, y, z]))
```

```python
bpy.ops.mesh.primitive_uv_sphere_add(radius=0.25, location=(1.5, 0.0, 0.5))
```
